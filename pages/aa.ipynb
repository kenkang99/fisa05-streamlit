```python
import requests
import pandas as pd

url = 'https://www.work24.go.kr/cm/openApi/call/hr/callOpenApiSvcInfo310L02.do'
# 반복해서 모든 페이지 데이터 가져오기
all_data = []
page = 1
i=0
while i <= len(df):
    params = {
        "authKey": "4a606afd-193d-4881-a83d-510c5d586276",
        "returnType": "JSON",
        "outType": "1",
        'srchTrprId' : df['trprId'][i],
        'srchTrprDegr' : "",
        "srchTorgId": str(df['trainstCstId'][i])
    }
    response = requests.get(url, params=params)
    

    try:
        data = response.json()
    except Exception as e:
        print("JSONDecodeError 발생:", e)
        print("응답 내용:", response.text) 
        break

#    print(data)

    # inst_base_info, inst_detail_info 중 하나라도 값이 있으면 all_data에 추가
    for key in ['inst_base_info', 'inst_detail_info']:
        if key in data and data[key]:
            all_data.append(data[key])

    i += 1

# 전체 데이터프레임으로 병합
df1 = pd.DataFrame(all_data)
df1
```

```python
import pandas as pd

df1=pd.read_csv("../훈련과정_전체데이터_ver3.csv")
df1
```

```python
df1.columns
```

```python
df1_unique=df1[~df1['제목'].duplicated()]
df1_unique['NCS_코드']=df1_unique['NCS_코드'].astype(str)
```

```python

df1_unique[df1_unique.loc[df1_unique['NCS_코드'].str[:2] == '20', '1차분류'] == '정보통신'
        
```

```python

df1_unique[df1_unique['1차분류'] == '정보통신']
```

```python
df1_unique.drop(columns=['2차분류', '3차분류'], inplace=True)
```

```python
df1_unique.loc[
    (df1_unique['1차분류'] == '정보통신') & (df1_unique['NCS_코드'].str[2:4] == '01'),
    '2차분류'
] = '정보기술'

df1_unique.loc[
    (df1_unique['1차분류'] == '정보통신') & (df1_unique['NCS_코드'].str[2:4] == '02'),
    '2차분류'
] = '통신기술'

df1_unique.loc[
    (df1_unique['1차분류'] == '정보통신') & (df1_unique['NCS_코드'].str[2:4] == '03'),
    '2차분류'
] = '방송기술'
```

```python
df1_unique
```

```python
mapping = {
    '01': '정보기술전략·계획',
    '02': '정보기술개발',
    '03': '정보기술운영',
    '04': '정보기술관리',
    '05': '정보기술영업',
    '06': '정보보호',
    '07': '인공지능',
    '08': '블록체인',
    '09': '스마트물류',
    '10': '디지털트윈',
    '11': '개인정보보호'
}

mask = df1_unique['2차분류'] == '정보기술'
df1_unique.loc[mask, '3차분류'] = df1_unique.loc[mask, 'NCS_코드'].str[4:6].map(mapping)
```

```python
df1_unique[mask]
```

```python
df1_unique[mask]['3차분류'].value_counts()
```

```python
df1_unique.to_csv("../훈련과정_전체데이터_ver4.csv", index=False)
```

```python
import plotly.express as px
import warnings

warnings.filterwarnings('ignore')

counts = df1_unique[mask]['3차분류'].value_counts()
fig = px.bar(
    counts,
    x=counts.index,
    y=counts.values,
    labels={'x': '3차분류', 'y': '훈련 개수'},
    title='3차분류별 훈련 개수'
)
fig.update_xaxes(tickangle=45)
fig.show()
```
